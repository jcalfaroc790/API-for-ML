```python
from faker import Faker
import pandas as pd
```

```python
#Creating Data
# Start Faker y define number of samples
fake = Faker()
num_samples = 1000

# Generate sinthetic data
data = {
    "mentor_domain_id": [fake.random_int(min=1, max=10) for _ in range(num_samples)],
    "mentor_id": [fake.random_int(min=1, max=30000) for _ in range(num_samples)],
    "reg_date_mentor": [fake.date_time_this_year() for _ in range(num_samples)],
    "region_id_mentor": [fake.random_int(min=1, max=10) for _ in range(num_samples)],
    "mentee_id": [fake.random_int(min=1, max=30000) for _ in range(num_samples)],
    "region_id_mentee": [fake.random_int(min=0, max=10) for _ in range(num_samples)],
    "session_id": [fake.random_int(min=1000, max=9999) for _ in range(num_samples)],
    "session_date_time": [fake.date_time_this_decade() for _ in range(num_samples)],
    "session_status": [fake.random_element(elements=("finished", "canceled")) for _ in range(num_samples)]
}

# Create a DataFrame with the data generated con los datos generados
synthetic_df = pd.DataFrame(data)

# Save the DataFrame in a CSV file
synthetic_df.to_csv("synthetic_data.csv", index=False)

# Show a previous view of the file
print(synthetic_df.head())
```
